```python
import os
```

```python
lang = "annotation_nynorsk"
finished = "FINISHED_ANNOTATIONS_NYNORSK"
path = os.path.join(os.getcwd(), lang, finished)
```

```python
from collections import defaultdict
files_with_dir = defaultdict(list)
# walk through the directory and get all the files
# add their paths to the dictionary

for root, dirs, files in os.walk(path):
    for file in files:
        if ".ann" in file: # disregard txts, assuming 1 to 1
            # only append the path from the "FINISHED_ANNOTATION" folder:
            short_path = root.split(finished)[1]
            files_with_dir[file].append(short_path)
```

```python
import json
# save the dictionary to a json file
with open("finished_nynorsk.json", "w") as f:
    json.dump(files_with_dir, f, indent=4)
```

# Process the data

```python
# nynorsk
first_filter = defaultdict(list)
for fileid, paths in files_with_dir.items():
    if len(paths) > 1:
        for p in paths:
            if "review" in p:
                first_filter[fileid].append(fileid)
            elif "IAA_" in p:
                # if there are multiple IAA folders, we want the one with the highest number
                # format: NN_RUNDEX_XX_IAA_A_user
                # or: NN_RUNDEX_XX_IAA_B_user
                # select B > A:
                if "IAA_B" in p:
                    first_filter[fileid].append(fileid)
    else:
        first_filter[fileid].append(fileid)

for fileid, paths in first_filter.items():
    print(fileid, paths)
```
